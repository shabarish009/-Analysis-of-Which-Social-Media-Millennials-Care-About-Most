```python
# Plot count plot graph
values = {'Count':5423}
df1 = df.fillna(value=values)

plt.figure(figsize=(24,8))
ax = sns.countplot(data = df, x='Segment Type')

plt.title('Segment Count by Type', c='black', loc='center')
plt.xticks(rotation=90)

for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()+2))
```

As per count plot, University segment has highest number of responses 308.
Followed by Custome 192.

```python
#Plot bar graph
values = {'Count':5423}
df1 = df.fillna(value=values)

plt.figure(figsize=(24,8))
ax = sns.barplot(data = df, x='Answer', y='Count', color='purple')

plt.title('Segment Count Segment Description', c='black', loc='center')
plt.xticks(rotation=90)

for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.3, p.get_height()+2))
```

As seen above: 
1. Snapchat has highest count 164
2. Followed by instagram with 79.3

Graph shows that Snapchat is most popular preference amonth others.
